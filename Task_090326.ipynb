```python
#1. Given a list of numbers, create a function to find the sum of all positive numbers

def sum(list1):
    total = 0
    for num in list1:
        if num>0:
            total+= num
    return total

sum([1,-3,-2,3,4,-9])
```

```python
#2. Write a python function to check if a given string is a palindrome

def isPalindrome(str):
    str = str.lower()
    if str == str[::-1]:
        return "Its a Palindrome"
    else:
        return "Its not a Palindrome"
        
    
print(isPalindrome(input("Enter a text to check Palindrome: ")))
```

```python
numbers = [1,2,3,4,5,6]

def square_list(numbers):
    squares = []
    for num in numbers:
        squares.append(num**2)

    return squares

print(square_list(numbers)) 
```
