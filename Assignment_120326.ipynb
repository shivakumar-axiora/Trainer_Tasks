```python
# 1. second largest element in list without using sort()

num = [1,2,3,4,5,6]

largest = 0 
second = 0

for n in num:
    if n > largest:
        second = largest
        largest = n
    elif n > second:
        second = n

print("Second Largest element in List: ",second)


```

```python
# 2. Find All Pairs with Given Sum. Return all pairs whose sum equals a given target.

nums = [1,2,3,4,5]

total = 5
def find_pairs(nums, total):
    pairs = []

    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == total:
                pairs.append((nums[i], nums[j]))

    return pairs

print(find_pairs(nums, total))
```

```python
# 3.Python program to Move all zeros to the end without changing order of other elements in a list.

numbers = [0,2,0,0,3,4,5]

new_list = []
zero_count = 0

for num in numbers:
    if num == 0:
        zero_count += 1
    else:
        new_list.append(num)

new_list.extend([0] * zero_count)
new_list
```

```python
# 4.Group words that are anagrams.

words = ["eat","tea","tan","ate","nat","bat"]

def group_anagrams(words):
    result = {}
    
    for word in words:
        key = "".join(sorted(word))
        result.setdefault(key, []).append(word)
        
    return list(result.values())

print(group_anagrams(words))
```

```python
# 5. Merge two dictionaries. If key exists, add values.

d1 = {'a':10,'b':20}
d2 = {'b':5,'c':15}

d3 = d1.copy()

for key,value in d2.items():
    if key in d3:
        d3[key] += d2[key]
    else:
        d3[key] = value
print("Merged Dictionary: ",d3)
```

```python
# 6. Find the longest words

words = ["python","machine","AI","data"]

longest_word = ''

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print(longest_word)
```

```python
#7. NUmber of Vowels and Consonants in a string

text = input("Enter any text: ")

vowels = 'aeiouAEIOU'
vowels_count = 0
consonants_count = 0

for t in text:
    if t in vowels:
        vowels_count += 1
    else:
        consonants_count += 1

print("Vowels: ", vowels_count)
print("Consonants: ", consonants_count)


```

```python
# 9. Compress a String 

s = "shhiiivvaaa"
compressed = ""

count = 1

for i in range(len(s)):
    if i < len(s)-1 and s[i] == s[i+1]:
        count += 1
    else:
        compressed += s[i] + str(count)
        count = 1

print(compressed)
```

```python
# 10. Generate all permutations of a string.

def permutations(s, ans=""):
    
    if len(s) == 0:
        print(ans, end=" ")
        return
    
    for i in range(len(s)):
        ch = s[i]
        remaining = s[:i] + s[i+1:]
        
        permutations(remaining, ans + ch)


string = "abc"
permutations(string)
```
